<h3 style="color:blue">Step 5 : Perform various charting and not down observations</h3>

```python
labels = ['Dibetic','Non-Diabetic']
colors = ['red','green']
sizes = dataSet.Outcome.value_counts().values
py.figure(figsize=(7,7))
py.pie(sizes,labels = labels,colors=colors,autopct='%1.1f%%')
          
```

<ul style="color:red">
    <li><h3>From the dataset ,Approx. 65% of the Womens are having diabetes and 35% does not have diabetes</h3></li>
 </ul>

```python
positiveDataSet = dataSet[dataSet.Outcome==1]
negativeDataSet = dataSet[dataSet.Outcome == 0]
```

```python
positiveDataSet.describe()
```

```python
negativeDataSet.describe()
```

<h3 style="color:red">Observation :</h3>
<ul>
    <li><h3 style="color:red">Glucose level in Diabetic womens are quire high as Mean of Diabetic patient is 141 whereas for non-diabetic its 109. Clearly High Glucose attributing to the Outcome.</h3></li>
    <li><h3 style="color:red">Mean age in Positive case is around 38 which clearly indicates that Diabetes probability increase with increase in Age. Diabetes mostly not observed in low age group people.</h3></li>
    <li><h3 style="color:red">Just like Glucose, Insulin also has been found to be on higher side for the people having Diabetes. With the mean of 100 and 68 in Positive and Negative cases respectively clearly indicates that Diabetic patients do have high level of Insulin </h3></li>
</ul>
    

```python
sb.distplot(positiveDataSet.Glucose)
sb.distplot(negativeDataSet.Glucose)
```

<h3 style="color:red">Observation :</h3>

    <li><h3 style="color:red">It can be observed here that with increased level of Glucose, chances of Diabetes increases</h3></li>
</ul>
